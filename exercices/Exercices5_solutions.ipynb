<u>Exercice 1:</u> *Déclaration et valeurs.*  
  
1- Déclarer un tableau numpy de 3 colonnes et 4 lignes, rempli de valeurs entières aléatoires entre 0 et 50  
2- Convertir les éléments de ce tableau en nombres flottants codés sur 16 bits  
3- Afficher les informations de forme et de dimensions du tableau  
4- Multiplier les valeurs de éléments du tableau par 2  
5- Donner la moyenne des valeurs des éléments du tableau et leur écart type  
6- Changer les dimensions du tableau pour avoir 2 lignes et 6 colonnes   

 ***<center>SOLUTION</center>***

```python
import numpy as np
```

```python
# 1- Déclarer un tableau numpy de 3 colonnes et 4 lignes, rempli 
# de valeurs aléatoires entre 0 et 50
tableau = np.random.randint(0, 50, size=(3, 4))
print('tableau déclaré:\n',tableau)
```

```python
# 2- Convertir les éléments de ce tableau en nombres 
# flottants codés sur 16 bits
tableau =  tableau.astype(np.float16)
print('tableau float 16 bits\n',tableau)
print('type du premier élément: ',type(tableau[0,0]))
```

```python
# 3- Afficher les informations de forme et de dimensions du tableau
print("Dimensions du tableau :", tableau.shape)
print("Nombre de dimensions du tableau :", tableau.ndim)
```

```python
# 4- Multiplier les valeurs de éléments du tableau par 2 
tableau  = tableau*2
print(tableau)
```

```python
# 5- Donner la moyenne des valeurs des éléments du tableau et leur écart type  
moyenne = np.mean(tableau)
print("Moyenne des valeurs :", moyenne)
ecart_type = np.std(tableau)
print("Ecart type des valeurs :", ecart_type)
```

```python
# 6- Changer les dimensions du tableau pour avoir 2 lignes et 6 colonnes 
tableau = tableau.reshape(2, 6)
print("Tableau après modification des dimensions :\n", tableau)
```

<u>Exercice 2:</u> *Indexation et conditions*  
  
1- Déclarer un tableau numpy de 4 colonnes et 4 lignes, rempli de valeurs décimales aléatoires entre 0 et 100  
2- afficher le tableau avec 1 décimale  
3- Extraire les valeurs de la première ligne du tableau et les assigner à une nouvelle variable appelée "sous_tableau". Afficher le contenu du sous-tableau extrait.   
4- Même question pour la dernière colonne du tableau
<div style="background-color: lightgrey; ">
5- Calculer la somme des éléments de chaque colonne du tableau d'origine</br>    
6- Donner le numéro de ligne et de colonne de chaque élement dont la valeur est supérieure à 50  
</div>

 ***<center>SOLUTION</center>***

```python
import numpy as np
```

```python
# 1- Déclarer un tableau numpy de 4 colonnes et 4 lignes, 
# rempli de valeurs décimales aléatoires entre 0 et 100
tableau = np.random.uniform(0, 100, size=(4, 4))
print('Tableau origine:\n',tableau)
```

```python
# 2- afficher le tableau avec 1 décimale
tableau = np.round(tableau,1)
print('\nTableau avec 1 décimale:\n',tableau)
```

```python
# 3- Extraire les valeurs de la première ligne du tableau et les assigner à 
# une nouvelle variable appelée "sous_tableau". 
# Afficher le contenu du sous-tableau extrait.

#Option1
sous_tableau = tableau[0,:]
print("Premiere ligne :\n", sous_tableau)
```

```python
# 4- Même question pour la dernière colonne du tableau

#Option1
sous_tableau = tableau[:,-1]
print("Derniere colonne :\n", sous_tableau)
```

```python
# 5 sup- Calculer la somme des éléments de chaque colonne du tableau d'origine 
somme_col = np.sum(tableau,axis = 0)
print('Somme des elements de chaque colonne',somme_col)
```

```python
# 6 sup- Donner le numéro de ligne et de colonne de chaque élement dont la valeur 
# est supérieure à 50  
indices_elements_sup50 = np.where(tableau>50)
print('Indices des elements superieurs à 50:')
print(indices_elements_sup50)

valeurs_elements_sup50 = tableau[indices_elements_sup50]
print('Valeur des elements superieurs à 50:')
print(valeurs_elements_sup50)
```

Ici on a recupéré un tuple indices_elements_sup50 avec en position 0 les lignes et en position 1 les colonnes, nous allons en extraire les numeros de ligne et de colonne

```python
indices_elements_sup50 = np.where(tableau > 50)

# Afficher les indices et les valeurs des éléments supérieurs à 50
print("Indices et valeurs des éléments supérieurs à 50 :")
for i in range(len(indices_elements_sup50[0])):
    ligne = indices_elements_sup50[0][i]
    colonne = indices_elements_sup50[1][i]
    valeur = tableau[ligne, colonne]
    print(f"Element {i+1} : Ligne {ligne}, Colonne {colonne}, Valeur {valeur}")
```

<u>Exercice sup: </u> *interpolation et dérivée*
  
1- créer un tableau numpy appelé X de dimension 1, contenant des valeurs croissantes de 1 à 10 par pas de 1.  
2- créer un tableau numpy appelé Y de dimension 1, contenant le même nombre d'éléments que X avec des valeurs aléatoires entre 0 et 10.  
3- Représenter Y en fonction de X sur un graphique.  
4- Définir un tableau Y2 contenant les valeurs de Y pour X2 = 0.5, X = 1.5, ....X = 9.5 avec une interpolation.  
5- Représenter les points de coordonnées X2,Y2 sur le  graphique avec des marqueurs de couleurs et forme differents de X,Y.  
6- Fusionner d'abord les tableaux X et X2 puis Y et Y2, c'est à dire créer un tableau dont les indices pairs auront les valeurs de X et les indices impairs celles de X2. Idem pour Y et Y2.    
On commencera par créer un tableau vide que l'on remplira ensuite.  
7- Vérifier en faisant le graphique des tableaux fusionnés.   
8- Tracer la dérivée des Y fusionnés par rapport aux X fusionnés. Représenter la derivée nulle par une ligne horizontale à y = 0  
9- Donner l'ordonnée du point 0.25 par une interpolation.  
</div>

 ***<center>SOLUTION</center>***

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1- creer un tableau numpy appelé X de dimension 1, contenant 
# des valeurs croissantes de 1 à 10 par pas de 1.

# Option 1 avec np.arange
X = np.arange(11)
print('X:\n',X)
```

```python
# Option 1 avec np.linspace
X = np.linspace(0, 10, 11, dtype=int)
print('X:\n',X)
```

```python
# 2- creer un tableau numpy appelé Y de dimension 1, 
# contenant le même nombre d'éléments que X avec des valeurs 
# aléatoires entre 0 et 10.
taille = len(X)
Y = np.random.randint(0, 10, size=(taille))
print('Y:\n',Y)
```

```python
# 3- Représenter Y en fonction de X sur un graphique.
fig,ax = plt.subplots()
ax.plot(X,Y,marker = 'o',markeredgecolor='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
# 4- Définir un tableau Y2 contenant les valeurs de Y pour 
# X2 = 0.5, X = 1.5, ....X = 9.5 avec une interpolation.
X2 = np.arange(0.5,10.5,1)
print('X2:\n',X2)
Y2 = np.interp(X2, X, Y)
print('Y2:\n',Y2)
```

```python
# 5- Représenter les points de coordonnées X2,Y2 sur le graphique 
# avec des marqueurs de couleurs et forme differents de X,Y.
fig,ax = plt.subplots()
ax.plot(X,Y,marker = 'o',markeredgecolor='blue')
ax.scatter(X2,Y2,marker = '*',color='red')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xticks(np.arange(0, 10, 1));
```

```python
# 6- Fusionner d'abord les tableaux X et X2 puis Y et Y2, c'est à dire créer 
# un tableau dont les indices pairs auront les valeurs de X et les indices impairs 
# celles de X2. Idem pour Y et Y2.    
# On commencera par créer un tableau vide que l'on remplira ensuite.  

# creer un tableau vide de la taille de X + taille de X2
# du même type que X2 (pour avoir des décimales)
fusionX =np.empty(X.shape[0] + X2.shape[0], dtype=X2.dtype)
# remplir ce tableau en partant de l'indice 0 et en incrementant 
# l'indice de (indices pairs)
# debut = indice 0, fin non specifiée (jusqu'à la fin), pas = 2
fusionX[0::2] = X 
# remplir ce tableau en partant de l'indice 1 et en incrementant l'indice de (indices impairs)
# debut = indice 1, fin non specifiée (jusqu'à la fin), pas = 2
fusionX[1::2] = X2 
# Fusion des Y
fusionY =np.empty(Y.shape[0] + Y2.shape[0], dtype=X2.dtype)
                                                        
fusionY[0::2] = Y 
fusionY[1::2] = Y2 

print('------------fusionX------------------')
print(fusionX)
print('------------fusionY------------------')
print(fusionY)
```

Remarque: Ici on a utilisé X.shape[0] , on aurait pu utiliser X.size qui est équivalent dans le cas d'un tableau 1D.

```python
# 7- Vérifier en faisant le graphique des tableaux fusionnés
fig,ax = plt.subplots()
ax.plot(fusionX,fusionY, label = 'Fusion')
ax.scatter(X,Y,marker = '+',c='orange',s = 100, label = 'X Y')
ax.scatter(X2,Y2,marker = '*',c='green',s = 100, label = 'X2 Y2')
ax.legend()
;
```

```python
# 8- Tracer la dérivée des Y fusionnés par rapport aux X fusionnés.  Représenter la derivée nulle 
# par une ligne horizontale à y = 0  

# Calcul de la dérivée de y par rapport à x
derivee = np.gradient(fusionY, fusionX)
fig,ax = plt.subplots()
ax.plot(fusionX,fusionY,marker = 'o',markeredgecolor='blue',label = 'Fusion')
ax.plot(fusionX,derivee,label = 'Derivee')
ax.axhline(y= 0, color='grey', linestyle='--')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Fusion et derivee')
ax.legend();
```

```python
#9- Donner l'ordonnée du point 0.25 par une interpolation. 
ordonnee_point_025 = np.interp(0.25, fusionX, fusionY)
print("L'ordonnée du point 0.25 est :", ordonnee_point_025)
```
