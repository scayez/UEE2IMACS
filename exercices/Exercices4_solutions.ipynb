<u>Exercice7:</u> *Puissance*  
  
1- Ecrire une fonction 'liste_croissante' qui crée une liste d'éléments croissants prenant en argument debut, fin, pas. On utilisera une boucle for pour générer cette liste.  
2- Ecrire une fonction 'puissance' qui éléve à un puissance passée en argument les éléments d'une liste, également passée en argument.  
3- Ecrire une fonction 'graphique' qui ajoute a une figure et axes existant un tracé. C'est a dire que la fonction prendra en argument, non seulement las valeurs à tracer, mais aussi la figure et l'axe, ainsi que la puissance à laquelle est élevée la liste.  
4- Utiliser, dans une boucle, les 3 fonctions définies précedemment pour:  
    - Créer une liste de nombres de 1 à 100 avec un pas de 2  
    - Mettre successivement cette liste à l'exposant 1 à 10  
    - Tracer sur un même graphe une représentation des valeurs de la liste aux exposants de 1 à 10    
5- Ce graphe est il facilement lisible? Comment l'améliorer?  

 ***<center>SOLUTION</center>***

```python
# 1- Ecrire une fonction 'liste_croissante' qui crée une liste d'éléments croissants 
#prenant en argument debut, fin, pas. On utilisera une boucle for pour générer cette liste.
def liste_croissante(debut, fin, pas):
    liste = []
    for i in range (debut, fin, pas):
        liste.append(i)
    return liste
```

```python
# 2- Ecrire une fonction 'puissance' qui éléve à un puissance passée en argument 
# les éléments d'une liste, également passée en argument.
def puissance(liste,exposant):
    puissance_liste = []
    for element in liste:     
        puissance_exposant_de_nombre = element**exposant
        puissance_liste.append(puissance_exposant_de_nombre)
    return puissance_liste
```

```python
# 3- Ecrire une fonction 'graphique' qui ajoute a une figure et axes existant un tracé. 
# C'est a dire que la fonction prendra en argument, non seulement las valeurs à tracer, 
# mais aussi la figure et l'axe, ainsi que la puissance à laquelle est élevée la liste.

def graphique(liste,puissance_liste,figure,axe,puissance):
    ax.plot(liste,puissance_liste,label = str(puissance))
    ax.legend()
    ax.set_xlabel('Valeur')
    ax.set_ylabel('Puissance de Valeur')
```

```python
# 4- Utiliser, dans une boucle, les 3 fonctions définies précedemment pour:
# - Créer une liste de nombres de 1 à 100 avec un pas de 2
# - Mettre successivement cette liste à l'exposant 1 à 10 
# - Tracer sur un même graphe une représentation des valeurs de la liste aux exposants de 1 à 10
fig,ax = plt.subplots()
for n in range (0,10):
    a = liste_croissante(1,100,2)
    a_puissance_n= puissance(a,n)
    graphique(a,a_puissance_n,fig,ax,n)
```

```python
# 5- Ce graphe est il facilement lisible? Comment l'améliorer? 

# --> Mettre l'axe des y en echelle log

fig,ax = plt.subplots()
for n in range (0,10,1):
    a = liste_croissante(1,100,2)
    a_puissance_n= puissance(a,n)
    graphique(a,a_puissance_n,fig,ax,n)
ax.set_yscale('log')
```
